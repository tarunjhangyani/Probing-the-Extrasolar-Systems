```python
from math import sqrt
import pandas as pd
# Want to display all the columns of the data frames, the below command will ensure that.
pd.set_option('display.max_columns', None)
from matplotlib import pyplot as plt
from astropy.constants import M_sun, R_sun, M_jup, au
n = 2  # Decide the no.of rows that are sufficient to display the data without clutter.
```

```python
# Define a function that takes in x and y data as argument along with other parameters and results out the scatter plots of the data.
def plotting(x, y, text, xtext, ytext, title, x_label, y_label):
    for i in range(len(x)):
        Xdata = x.iloc[i]
        Ydata = y.iloc[i]
        Text = text.iloc[i]
        plt.scatter(Xdata,Ydata)
        plt.text(Xdata + xtext, Ydata + ytext, Text)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid()
    plt.show()
```

```python
def habitablezoneCount(Luminosity, distance_series):
    Absolute_luminosity = 10 ** (Luminosity)
    rinner = sqrt(Absolute_luminosity / 1.1)
    router = sqrt(Absolute_luminosity / 0.53)

    count = 0
    for i in range(0, len(distance_series)):
        if rinner <= distance_series.iloc[i] <= router:
            count = count + 1
    print('The inner limit of habitable zone is:', rinner,'AU')
    print('The outer limit of habitable zone is:', router,'AU')
    return count
```

# Analysis of Solar System Data

```python
# Reading the csv file containing Solar System oblects data.
solar_df = pd.read_csv('/content/drive/MyDrive/Projects/Sem 2/Astro/CSVs/Solar System data.csv')

# Display few rows for better readability
print('Original data set of solar system oblecjs', 'displaying only', n, 'rows')
display(solar_df.head(n))
print('\n')

# Sorting out only Planets.
condition = solar_df['isPlanet'] == True  # This condition is a boolean series.
planets_df = solar_df[condition]          # Applying the condition on data frame gives dataframe of planets only with no change in the orginal data frame.

# Display the planets data frame (planets_df).
print(f'Data frame of planets in our solar system and there are {len(planets_df)} planets in our solar system according to csv input')
display(planets_df)
print('\n')

# Drop the planets other than 8 known planets.
planets_df = planets_df.drop([179, 187, 208, 222, 236], axis = 0)
# Create subset of planets_df with relevant columns only
planets_df = planets_df[['eName', 'eccentricity', 'massj', 'semimajorAxis_AU']]
print('This is the required subset of the planets_df with information of relevant rows and columns only')
display(planets_df)


print('The number of planets in habitable zone in our solar system are:', habitablezoneCount(0, planets_df['semimajorAxis_AU']))

# Plot the relevant graphs.
# 1) Mass distribution (mass vs semi major axis)
for i in range(len(planets_df['eName'])):
    plt.scatter(planets_df['semimajorAxis_AU'].iloc[i], planets_df['massj'].iloc[i], label = planets_df['eName'].iloc[i])
plt.legend()
plt.title('Mass distribution')
plt.xlabel('Distance from sun (AU)')
plt.ylabel('Mass of planets (jupiter mass)')
plt.grid()
plt.show()

# 1) Eccentricity distribution (eccentricity vs distribution)
for i in range(len(planets_df['eName'])):
    plt.scatter(planets_df['semimajorAxis_AU'].iloc[i], planets_df['eccentricity'].iloc[i], label = planets_df['eName'].iloc[i])
plt.legend()
plt.title('Eccentricity distribution')
plt.xlabel('Distance from sun (AU)')
plt.ylabel('Eccentricity')
plt.grid()
plt.show()

# Normalize the planets_df.
planets_df['massj'] = planets_df['massj'] * (M_jup / M_sun)
planets_df['semimajorAxis_AU'] = planets_df['semimajorAxis_AU'] * (au / R_sun)
# Give the columns appropriate names
planets_df.rename(columns = {'massj':'mass(solar mass)', 'semimajorAxis_AU': 'semi-major axis(solar radius)'}, inplace = True)
print('Planets data after normalization')
display(planets_df)

# Plot the relevant graphs.
# 1) Mass distribution (mass vs semi major axis)
for i in range(len(planets_df['eName'])):
    plt.scatter(planets_df['semi-major axis(solar radius)'].iloc[i], planets_df['mass(solar mass)'].iloc[i], label = planets_df['eName'].iloc[i])
plt.legend()
plt.title('Mass distribution')
plt.xlabel('Distance from sun (solar radius)')
plt.ylabel('Mass of planets (solar mass)')
plt.grid()
plt.show()

# 1) Eccentricity distribution (eccentricity vs distribution)
for i in range(len(planets_df['eName'])):
    plt.scatter(planets_df['semi-major axis(solar radius)'].iloc[i], planets_df['eccentricity'].iloc[i], label = planets_df['eName'].iloc[i])
plt.legend()
plt.title('Eccentricity distribution')
plt.xlabel('Distance from sun (solar radius)')
plt.ylabel('Eccentricity')
plt.grid()
plt.show()
```

```python
# Reading the csv file containing exoplanet data
exo_df = pd.read_csv('/content/drive/MyDrive/Projects/Sem 2/Astro/CSVs/Exoplanets Data.csv')

# Rename all the columns with more meaning-full name.
newname_dict = {
                'loc_rowid':'Serial No.',
                'pl_name':'Planet Name',
                'hostname':'Star Name',
                'pl_letter':'Planet Letter',
                'sy_snum':'Number of Stars',
                'sy_pnum':'Number of Planets',
                'discoverymethod':'Discovery Method',
                'disc_year':'Discovery Year',
                'pl_controv_flag':'Controversial Flag',
                'pl_orbper':'Orbital Period [days]',
                'pl_orbsmax':'Orbit Semi-Major Axis [au]',
                'pl_rade':'Planet Radius [Earth Radius]',
                'pl_radj':'Planet Radius [Jupiter Radius]',
                'pl_bmasse':'Planet Mass or Mass*sin(i) [Earth Mass]',
                'pl_bmassj':'Planet Mass or Mass*sin(i) [Jupiter Mass]',
                'pl_bmassprov':'Planet Mass or Mass*sin(i) Provenance',
                'pl_orbeccen':'Eccentricity',
                'pl_insol':'Insolation Flux [Earth Flux]',
                'pl_eqt':'Equilibrium Temperature [K]',
                'pl_orbincl':'Inclination [deg]',
                'st_spectype':'Spectral Type',
                'st_teff':'Stellar Effective Temperature [K]',
                'st_rad':'Stellar Radius [Solar Radius]',
                'st_mass':'Stellar Mass [Solar mass]',
                'st_lum':'Stellar Luminosity [log(Solar)]',
                'sy_dist':'Distance [pc]',
                'sy_plx':'Parallax [mas]'
                }
exo_df = exo_df.rename(columns = newname_dict)


'''Some Analysis before moving to find planetary system'''
# Plotting Histogram of masses of planet (Earth Mass)
print('Histogram of masses of planets', '\n')
plt.hist(exo_df['Planet Mass or Mass*sin(i) [Earth Mass]'], color = 'dodgerblue')
plt.xlabel('Mass of planets (Earth mass)')
plt.ylabel('Frequency')
plt.title('Histogram of masses of planets (Earth Mass)')
plt.show()
# Plotting Narrow Histogram of masses of planet (Earth Mass)
print('Narrower Histogram of masses of planets', '\n')
condition_M = exo_df['Planet Mass or Mass*sin(i) [Earth Mass]'] <= 50
exo_M = exo_df['Planet Mass or Mass*sin(i) [Earth Mass]'][condition_M]
plt.hist(exo_M, color = 'dodgerblue',bins = 190)
plt.xlabel('Mass of planets (Earth mass)')
plt.ylabel('Frequency')
plt.title('Narrow Histogram of masses of planets (Earth Mass)')
plt.show()


# Search for earth 2.0 begins!!!

cond_EM = (exo_df['Planet Mass or Mass*sin(i) [Earth Mass]'] >= 0.8) & (exo_df['Planet Mass or Mass*sin(i) [Earth Mass]'] <= 1.2)
cond_ER = (exo_df['Planet Radius [Earth Radius]'] >= 0.8) & (exo_df['Planet Radius [Earth Radius]'] <= 1.2)
cond_ETP = (exo_df['Equilibrium Temperature [K]'] >= 250) & (exo_df['Equilibrium Temperature [K]'] <= 323)

Earthlike_condition = (cond_EM & cond_ER & cond_ETP)
Earth_2 = exo_df[Earthlike_condition]
Earth_2
drop_list = ['Planet Mass or Mass*sin(i) [Jupiter Mass]', 'Planet Mass or Mass*sin(i) Provenance', 'Stellar Radius [Solar Radius]', 'Serial No.','Parallax [mas]','Spectral Type','Planet Radius [Jupiter Radius]', 'Controversial Flag','Inclination [deg]', 'Number of Planets', 'Insolation Flux [Earth Flux]']
Earth_2 = Earth_2.drop(drop_list, axis = 1)
print('Yay!!!, we found Planet earth 2.0', '\n')
display(Earth_2)

print('The exoplanet data:','\n')
# Fill NaN(None) places with Unknown string to ease out (Not really!) the further processing of data
exo_df.fillna('Unknown', inplace = True)
# Display only the few rows for better readability
exo_df.head(n)
```

```python
# Define conditions to sort out the data.
sn = exo_df['Number of Stars'] == 1                            # This condition define number of stars in a particluar exostellar sytem.
pn = exo_df['Number of Planets'] >= 4                          # This condition define number of planets in a particluar exostellar sytem.
sp = exo_df['Spectral Type'] != 'Unknown'                      # This condition rejects the stars whose spectral type is unknown.
ec = exo_df['Eccentricity'] != 'Unknown'                       # This condition rejects the planets whose eccentricity is unknown.
sm = exo_df['Orbit Semi-Major Axis [au]'] != 'Unknown'         # This condition rejects the planets whose value of semi-major axis is unknown.

condition1 = (sn & pn & sp & ec & sm)                          # This statement (command) ensures that all the above conditions are True simultaneously.

# Apply the condition defined above to the exoplanet data exo_df).
print('This is a list of stars that meet the condition defined.', 2*'\n',exo_df['Star Name'][condition1].unique())
print('\n')
print('This is a list of spectral types of stars that meet the condition defined.', 2*'\n',exo_df['Spectral Type'][condition1].unique())
```
